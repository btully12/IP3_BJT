```python
def f_d(filename,x):
    import pandas as pd
    global df,df2,df3,df4 
    file=f'{filename}'
    if x==1: 
        df=pd.read_csv(f"{filename}.csv")
        df['Date']=pd.to_datetime(df['Date'],format='%d/%m/%Y')
        df.insert(0,"Company",file)
    elif x==2:
        df2=pd.read_csv(f"{filename}.csv")
        df2['Date']=pd.to_datetime(df['Date'],format='%d/%m/%Y')
        df2.insert(0,"Company",file)
    elif x==3:
        df3=pd.read_csv(f"{filename}.csv")
        df3['Date']=pd.to_datetime(df['Date'],format='%d/%m/%Y')
        df3.insert(0,"Company",file)
          
    else:
        df4=pd.read_csv(f"{filename}.csv")
        df4['Date']=pd.to_datetime(df['Date'],format='%d/%m/%Y')
        df4.insert(0,"Company",file)
        

f_d('50Hertz',1)
f_d('Amprion',2)
f_d('TransnetBW',3)
f_d('TenneTTSO',4)
```
